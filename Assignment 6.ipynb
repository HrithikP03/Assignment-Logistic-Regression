```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
#Load the dataset
data = pd.read_csv("bank-full.csv", sep=';')
```

```python
data.head()
```

```python
data.info()
```

```python
data.shape
```

```python
#One hot encoding of Categorical column
data1 = pd.get_dummies(data, columns=['job','marital','education','contact','month','poutcome'])
```

```python
data1
```

```python
#To see all the hidden columns
pd.set_option("display.max.columns", None)
data1
```

```python
data1.info()
```

```python
import numpy as np
```

```python
data1['default']= np.where(data1['default'].str.contains('yes'),1,0)
data1['housing']= np.where(data1['housing'].str.contains('yes'),1,0)
data1['loan']= np.where(data1['loan'].str.contains('yes'),1,0)
data1['y']= np.where(data1['y'].str.contains('yes'),1,0)
data1
```

Model Building

```python
#Dividing data into input and otput
x = pd.concat([data1.iloc[:,0:10], data1.iloc[:,11:]], axis=1)
y = data1.iloc[:,10]
```

```python
x
```

```python
y
```

```python
# Logistic regression model
classifier=LogisticRegression()
classifier.fit(x,y)
```

```python
#Predict for X dataset
y_pred = classifier.predict(x)
y_pred
```

```python
y_pred_df= pd.DataFrame({'Actual':y,'Predict': y_pred})
y_pred_df
```

```python
#Testing model accuracy
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y,y_pred)
print (confusion_matrix)
```

```python
# The model accuracy is calculated by (a+d)/(a+b+c+d)
((39162+1153)/(39162+760+4136+1153))*100
```

```python
#Classification report
from sklearn.metrics import classification_report
print(classification_report(y,y_pred))
```

ROC Curve

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

fpr, tpr, thresholds = roc_curve(y, classifier.predict_proba(x)[:,1])

auc = roc_auc_score(y, y_pred)

import matplotlib.pyplot as plt
plt.plot(fpr, tpr, color='red', label='logit model ( area  = %0.2f)'%auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
```
